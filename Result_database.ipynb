```python
database = {}

count = 1
serial_matric = 400000
from random import randrange
result = randrange(30,99) 

while count < 201:
    
    matric_no = 'UNN' + str(serial_matric)
    database[matric_no] = {'GST102':0, 'MKT101':0, 'CPT100':0}
    
  
    
    if result >=40 and result <=44:
        score_result = 'E'
       
    elif result >=45 and result <=49:
        score_result = 'D'
        
    elif result >=50 and result <=59:
        score_result = 'C'
            
    elif result >=60 and result <=69:
        score_result = 'B'
           
    elif result >=70:
        score_result = 'A'
        
    for scores in database[matric_no]:
        database[matric_no][scores] = score_result
    
    count+=1
    serial_matric+=1
```

```python
name = str(input('What\'s your name?\n'))

matric_num = input('Enter your matric number:\n')
mnproper = matric_num.upper()

subject = input('Enter Subject:\n')
sbjproper = subject.upper()

if mnproper in database:
    print(f'Hello, {name} ({mnproper}). You got "{(database[mnproper][sbjproper])}" in {sbjproper}.')
```
